## INDICADORES AO LONGO DE 5 DÉCADAS

```python
df[df['year'].between(1953, 2007)] \
    .sort_values('year', ascending=False)
```

```python
df[df['year'].between(1953, 2007)] 

sns.barplot(data=df, x='continent', y='pop')

plt.title('Crescimento Populacional por Continente (1953-2007)')
plt.xlabel('Continente')
plt.ylabel('População')
plt.show()
```

```python
df[df['year'].between(1953, 2007)] 

sns.barplot(data=df, x='continent', y='gdpPercap')

plt.title('Crescimento do PIB por Continente (1953-2007)')
plt.xlabel('Continente')
plt.ylabel('PIB per Capita')
plt.show()
```

```python
df[df['year'].between(1953, 2007)] 

sns.barplot(data=df, x='continent', y='lifeExp')

plt.title('Crescimento da Longevidade por Continente (1953-2007)')
plt.xlabel('Continente')
plt.ylabel('Longevidade')
plt.show()
```

```python
dados = df[
    (df['continent'] == 'Asia') &
    (df['year'].between(1953, 2007))
]

plt.figure(figsize=(20,6))

sns.barplot(
    data=dados,
    x='country',
    y='pop',
    order=dados.sort_values('pop', ascending=False)['country']
)

plt.xticks(rotation=90)

plt.title('Países Asiáticos Mais Populosos (1953-2007)')
plt.xlabel('Países')
plt.ylabel('População')
plt.show()
```

```python
dados = df[
    (df['continent'] == 'Oceania') &
    (df['year'].between(1953, 2007))
]

dados_pie = dados.groupby('country')['pop'].sum()
plt.figure(figsize=(20,6))

plt.pie(
    dados_pie,
    labels=dados_pie.index,
    autopct='%1.1f%%'
)

plt.xticks(rotation=90)

plt.title('Países Oceânicos com Maior PIB (1953-2007)')
plt.show()
```

```python
dados = df[
    (df['continent'] == 'Oceania') &
    (df['year'].between(1953, 2007))
]

dados_pie = dados.groupby('country')['lifeExp'].sum()
plt.figure(figsize=(20,6))

plt.pie(
    dados_pie,
    labels=dados_pie.index,
    autopct='%1.1f%%'
)

plt.xticks(rotation=90)

plt.title('Países Oceânicos com Maior Longevidade (1953-2007)')
plt.show()
```

```python
dados = df[
    (df['continent'] == 'Europe') &
    (df['year'].between(1953, 2007))
]

plt.figure(figsize=(20,6))

sns.barplot(
    data=dados,
    x='country',
    y='lifeExp',
    order=dados.sort_values('lifeExp', ascending=False)['country']
)

plt.xticks(rotation=90)

plt.title('Países Europeus com Maior Longevidade (1953-2007)')
plt.xlabel('Países')
plt.ylabel('Longevidade')
plt.show()
```
